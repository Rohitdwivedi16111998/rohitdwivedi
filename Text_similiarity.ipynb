```python
for i in range(4023):    
    u1=set(filtered_sentence_text1[i])
    u2=set(filtered_sentence_text2[i])
    l1=[]
    l2=[]
    rvector = u1.union(u2)  
    for w in rvector: 
        if w in u1: l1.append(1) # create a vector 
        else: l1.append(0) 
        if w in u2: l2.append(1) 
        else: l2.append(0) 

    c=0

    # cosine formula  
    for i in range(len(rvector)): 
            c+= l1[i]*l2[i] 
    cosine = c / float((sum(l1)*sum(l2))**0.5) 
    print("similarity: ", cosine) 
    scores.append(cosine)
```
