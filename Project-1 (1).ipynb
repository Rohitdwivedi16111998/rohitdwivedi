```python
df0['Points']= df0['Points'].apply(pd.to_numeric)
```

```python
df0.nlargest(5,'Points', keep='first') #Print the list of teams which came Top 5 in terms of points 
```

```python
df0['GoalsFor']= df0['GoalsFor'].apply(pd.to_numeric)
df0['GoalsAgainst']= df0['GoalsAgainst'].apply(pd.to_numeric)
```

```python
def Goal_diff_count(df0):
     return df0.GoalsFor - df0.GoalsAgainst
```

```python
Goal_diff_count(df0)
```

```python
df0['Goal_difference_count']=Goal_diff_count(df0)
```

```python
df0
```

```python
df0['Goal_difference_count'].max()
```

```python
df0[df0.Goal_difference_count == df0.Goal_difference_count.max()]   #the team which has maximum goal difference.
```

```python
df0[df0.Goal_difference_count == df0.Goal_difference_count.abs().min()] ##the team which has minimum goal difference.
```

```python
df0['GamesWon']= df0['GamesWon'].apply(pd.to_numeric)
```

```python
df0['GamesPlayed']= df0['GamesPlayed'].apply(pd.to_numeric)
```

```python
df0['GamesWon']
```

```python
df0['Winning Percent'] = (df0['GamesWon'] / df0['GamesPlayed'])*100
```

```python
df0=df0.fillna(0)
```

```python
df0.nlargest(5,'Winning Percent', keep='first')    #Print the top 5 teams which has the highest Winning percentage
```

```python
df0['BestPosition']= df0['BestPosition'].apply(pd.to_numeric)
```
